```python
import plotly.graph_objects as go
from typing import List, Tuple, Dict
from collections import defaultdict

def plot_coverage_track(infile: str, ref: str, chr: str, threshold: float):
    """
    
    """
    def get_ref_sequence() -> str:
        with open(ref, "r") as r:
            for line in r:
                if (line.startswith(">")) & (line[1:].strip() == chr):
                    return next(r).strip()
            raise Exception(f"Sequence not found. Chromosome '{chr}' not found in '{ref}'.")

    def set_up_data() -> Tuple[Dict[str, int]]:
        above_thr = {"site": [],
                     "reads": [],
                     "a": [],
                     "c": [],
                     "g": [],
                     "t": [],
                     "del": [],
                     "ins": []}
        below_thr = {"site": [],
                    "reads": [],
                    "a": [],
                    "c": [],
                    "g": [],
                    "t": [],
                    "del": [],
                    "ins": []}

        def add_line(line: list[str], data_dict: dict):
            data_dict["site"].append(int(line[1]))
            data_dict["reads"].append(int(line[2]))
            data_dict["a"].append(int(line[5]))
            data_dict["c"].append(int(line[6]))
            data_dict["g"].append(int(line[7]))
            data_dict["t"].append(int(line[8]))
            data_dict["del"].append(int(line[9]))
            data_dict["ins"].append(int(line[10]))
            return data_dict
        
        with open(infile, "r") as file:
            next(file)
            for line in file:
                line = line.strip("\n").split("\t")
                if line[0] == chr:
                    perc_mis = float(line[17])

                    if perc_mis >= threshold:
                        above_thr = add_line(line, above_thr)
                    else:
                        below_thr = add_line(line, below_thr)
        
        return above_thr, below_thr
    
    
    def create_plot():
        def get_custom_data(chr: str, data_dict: dict) -> List[List[str|int]]:
            custom_data = [[chr, 
                            data_dict["reads"][i],
                            data_dict["a"][i],
                            data_dict["c"][i],
                            data_dict["g"][i],
                            data_dict["t"][i], 
                            data_dict["del"][i],
                            data_dict["ins"][i],
                            ref_seq[i]]
                            for i in range(len(data_dict["site"]))]    
            return custom_data

        hover_template_top = "%{customdata[0]}:%{x}<br>" + \
            "Total count: %{customdata[1]}<br>" + \
            "----------"

        hover_template_bottom = "----------<br>" + \
            "DEL: %{customdata[6]}<br>" + \
            "INS: %{customdata[7]}<br>" + \
            "----------<br>" + \
            "Ref: %{customdata[8]}<extra></extra>"

        hover_template = hover_template_top + \
            "<br>A: %{customdata[2]}<br>" + \
            "C: %{customdata[3]}<br>" + \
            "G: %{customdata[4]}<br>" + \
            "T: %{customdata[5]}<br>" + \
            hover_template_bottom
        
        # CREATE ALL TRACES

        hidden_trace_top = go.Bar(x = above_thr["site"], y = [0 for _ in above_thr["site"]],
                                  name = "", customdata = get_custom_data(chr, above_thr),
                                  hovertemplate = hover_template_top, marker_color = "white")

        acgt_traces = []
        colors = {"a": "green", "c": "blue", "g": "orange", "t": "red"}
        bar_bases = ["a", "c", "g", "t"]
        for base in bar_bases:
            trace = go.Bar(width=1, x = above_thr["site"], y = above_thr[base],
                           name = base.upper(), marker_color = colors[base])
            acgt_traces.append(trace)

        remaining_trace = go.Bar(width=1, x = above_thr["site"], y = above_thr["del"],
                                 name = "", marker_color = "grey", hoverinfo = "skip")

        hidden_trace_bottom = go.Bar(x = above_thr["site"], y = [0 for _ in above_thr["site"]],
                                     customdata = get_custom_data(chr, above_thr),
                                     hovertemplate = hover_template_bottom, marker_color = "white")

        nonerror_hidden_trace = go.Bar(x = below_thr["site"], y = [0 for _ in below_thr["site"]],
                                       customdata = get_custom_data(chr, below_thr),
                                       hovertemplate = hover_template, marker_color = "white",
                                       hoverlabel = {})

        nonerror_trace = go.Bar(width=1, x = below_thr["site"], y = below_thr["reads"], 
                                name = "Total count", marker_color = "grey", hoverlabel = {},
                                hoverinfo = "skip")

        all_traces = [hidden_trace_top] + acgt_traces + [remaining_trace, hidden_trace_bottom, nonerror_hidden_trace, nonerror_trace]

        fig = go.Figure()
        fig.update_xaxes(range=[1, len(ref_seq)])
        fig.add_traces(all_traces)

        fig.update_layout(
            template = "simple_white",
            title="Coverage track",
            xaxis_title='Site',
            yaxis_title='Count',
            barmode='stack',
            hovermode="x unified",
            showlegend=False,
            yaxis = dict(
                rangemode = "nonnegative"
            )
        )
        return fig

    ref_seq = get_ref_sequence()
    above_thr, below_thr = set_up_data()

    fig = create_plot()
    fig.show()
```

```python

def get_custom_data(chr: str, data_dict: dict) -> List[List[str|int]]:
    custom_data = [[chr, 
                    data_dict["reads"][i],
                    data_dict["a"][i],
                    data_dict["c"][i],
                    data_dict["g"][i],
                    data_dict["t"][i], 
                    data_dict["del"][i],
                    data_dict["ins"][i],
                    ref[i]]
                    for i in range(len(data_dict["site"]))]    
    return custom_data

hover_template_top = "%{customdata[0]}:%{x}<br>" + \
    "Total count: %{customdata[1]}<br>" + \
    "----------"

hover_template_bottom = "----------<br>" + \
    "DEL: %{customdata[6]}<br>" + \
    "INS: %{customdata[7]}<br>" + \
    "----------<br>" + \
    "Ref: %{customdata[8]}<extra></extra>"

hover_template = hover_template_top + \
    "<br>A: %{customdata[2]}<br>" + \
    "C: %{customdata[3]}<br>" + \
    "G: %{customdata[4]}<br>" + \
    "T: %{customdata[5]}<br>" + \
    hover_template_bottom

colors = {
    "a": "green",
    "c": "blue",
    "g": "orange",
    "t": "red"
    }

fig = go.Figure()
fig.update_xaxes(range=[1, len(ref)])

# Add coloured bars to the plot (where the threshold was surpassed)

# add bars for the count of the bases and add to figure
acgt_traces = []

hidden_trace_top = go.Bar(
    x = above_thr["site"],
    y = [0 for _ in above_thr["site"]],
    name = "",
    customdata = get_custom_data(chr, above_thr),
    hovertemplate = hover_template_top,
    marker_color = "white",    
    hoverlabel = {}
)
acgt_traces.append(hidden_trace_top)

bar_bases = ["a", "c", "g", "t"]
for base in bar_bases:
    trace = go.Bar(
        width=1,
        x = above_thr["site"],
        y = above_thr[base],
        name = base.upper(),
        marker_color = colors[base],    
        hoverlabel = {}
    )
    acgt_traces.append(trace)

# add the remaining read count to the traces (i.e. deletion count atm)
remaining_trace = go.Bar(
    width=1,
    x = above_thr["site"],
    y = above_thr["del"],
    name = "",
    marker_color = "grey",
    hoverlabel = {},
    hoverinfo = "skip",
)
acgt_traces.append(remaining_trace)

hidden_trace_bottom = go.Bar(
    x = above_thr["site"],
    y = [0 for _ in above_thr["site"]],
    customdata = get_custom_data(chr, above_thr),
    hovertemplate = hover_template_bottom,
    marker_color = "white",    
    hoverlabel = {}
)
acgt_traces.append(hidden_trace_bottom)

fig.add_traces(acgt_traces)

# Add grey bars to the plot (where the threshold was not surpassed)

nonerror_hidden_trace = go.Bar(
    x = below_thr["site"],
    y = [0 for _ in below_thr["site"]],
    customdata = get_custom_data(chr, below_thr),
    hovertemplate = hover_template,
    marker_color = "white",    
    hoverlabel = {}
)
nonerror_trace = go.Bar(
    width=1,
    x = below_thr["site"],
    y = below_thr["reads"],
    name = "Total count",
    marker_color = "grey",    
    hoverlabel = {},
    hoverinfo = "skip",
)
fig.add_traces([nonerror_hidden_trace, nonerror_trace])

fig.update_layout(
    template = "simple_white",
    title="Coverage track",
    xaxis_title='Site',
    yaxis_title='Count',
    barmode='stack',
    hovermode="x unified",
    showlegend=False,
    yaxis = dict(
        rangemode = "nonnegative"
    )
)

fig.show()
```
